## Load dataset 

This project develops a **machine learning regression model** to predict **lithium-ion battery capacity** using processed data from the [**CALCE dataset**](https://calce.umd.edu/battery-data). The model uses **cycle-level features** extracted from charging and discharging phases to estimate remaining capacity and **State of Health (SoH)**.

## Dataset Overview

The dataset includes data from two battery cell types:

| Cell Type | Description |
|---------|-----------|
| **CS2** | Standard cylindrical/pouch cells |
| **CX2** | Variant with potentially different chemistry or form factor |

### Data Contents (Per Cycle)

- Extracted Features from **charging** and **discharging** phases 
- Measured Capacity which is the actual capacity (Ah) at each cycle
- Nominal Capacity this is the original rated capacity (used for determing battery SoH (%)=`(measured_capacity / nominal_capacity) × 100`)

> **Target**: Predict `measured_capacity` using relevant features  

## Objective

Build a **robust regression model** that:
- Accurately predicts battery capacity at any cycle
- Identifies **key degradation indicators**
- Generalizes across **CS2 and CX2** cells
- Supports **early failure detection** and **RUL (Remaining Useful Life)** estimation
